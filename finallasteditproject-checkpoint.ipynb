```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

last_edit = pd.read_csv("last edit.csv")

last_edit
```

```python
# Set x to our input Temperature
x = last_edit['Total score']
# Set y to our output humidity, this is the value we trying to predict
y = last_edit['rating']

# Plot Humidity against Temperature
plt.scatter(x, y)
# Add labels to the graph
plt.xlabel('Total score')
plt.ylabel('rating')
```

```python
from sklearn.linear_model import LinearRegression

x = last_edit['Total score'].values.reshape(-1,1)
y = last_edit['Total score']

lr_model = LinearRegression()
lr_model.fit(x, y)
y_pred = lr_model.predict(x)

plt.scatter(x, y)
plt.xlabel('Total score')
plt.ylabel('Total score')

plt.plot(x, y_pred)
```

```python
theta_0 = lr_model.intercept_
theta_1 = lr_model.coef_
theta_0, theta_1
```

```python
y_pred = lr_model.predict(np.array([32]).reshape(1,1))
y_pred
```
